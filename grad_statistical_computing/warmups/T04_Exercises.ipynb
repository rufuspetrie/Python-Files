**1**. Read the `mtcars` data frame from https://bit.ly/2b72LNj to a `pandas` DataFrame. Find the mean `wt` and `mpg` for all cars grouped by the number of `gear`s.

```python
# Read dataframe from url
mtcars = pd.read_csv("https://bit.ly/2b72LNj")
mtcars.head()
```

```python
# Group by gears, generate summary statistics
mtcars.groupby("gear").mean()[["wt", "mpg"]]
```

**2**. Download the attention dataset from https://bit.ly/2B1iGZB, treating column 1 as the index, and answer the following:

1. How many rows and columns are there?
2. Drop the `solutions` column
3. Sort by `score` in descending order
4. Who is the subject with the highest score in divided attention?
5. Show the mean and standard deviations of the scores by subject and attention.
6. Find the difference between the scores for each subject when solution = 1 and when solution = 2.

```python
# Read dataframe from url
data = pd.read_csv("https://bit.ly/2B1iGZB")
data = data.set_index("Unnamed: 0")
data.head()
```

```python
# Rows and columns
data.shape
```

```python
# Drop solutions
data = data.drop(columns="solutions")
```

```python
# Sort by score
data = data.sort_values(by="score", ascending=False)
```

```python
# Highest score for divided attention
data[data.attention=="divided"]["score"].max()
```
