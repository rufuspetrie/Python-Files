```python
%matplotlib inline
import daft
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
pgm = daft.PGM(shape=[3.5, 2.5], origin=[0, 0])

pgm.add_node(daft.Node("alpha", r"$\alpha$", 0.5, 0.5))
pgm.add_node(daft.Node("Z", r"$Z$", 1.5, 0.5))
pgm.add_node(daft.Node("sigma_A", r"$\sigma_A$", 0.5, 1.5))
pgm.add_node(daft.Node("A", r"$A$", 1.5, 1.5))
pgm.add_node(daft.Node("X", r"$X$", 3.0, 1.0))
pgm.add_node(daft.Node("sigma_X", r"$\sigma_X$", 2.25, 2.0))

pgm.add_edge("alpha", "Z")
pgm.add_edge("sigma_A", "A")
pgm.add_edge("Z", "X")
pgm.add_edge("A", "X")
pgm.add_edge("sigma_X", "X")

pgm.render()
plt.close()
pgm.figure.savefig("pgm.png", dpi=300)
Image("pgm.png", width=400)
```
