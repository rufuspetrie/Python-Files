```python
%matplotlib inline
import daft
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
pgm = daft.PGM(shape=[3.0, 1.0], origin=[0, 0])

pgm.add_node(daft.Node("alpha", r"$\alpha$", 0.5, 0.5))
pgm.add_node(daft.Node("pi_k", r"$\pi_k$", 1.5, 0.5))
pgm.add_node(daft.Node("z_ik", r"$z_{ik}$", 2.5, 0.5))

pgm.add_edge("alpha", "pi_k")
pgm.add_edge("pi_k", "z_ik")

pgm.render()
plt.close()
pgm.figure.savefig("IBP_pgm.png", dpi=300)
Image("IBP_pgm.png", width=400)
```
