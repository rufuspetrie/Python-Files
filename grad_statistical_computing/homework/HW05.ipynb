This exercise is about using Newton's method to find the cube roots of unity - find $z$ such that $z^3 = 1$. From the fundamental theorem of algebra, we know there must be exactly 3 complex roots since this is a degree 3 polynomial.

We start with Euler's equation
$$
e^{ix} = \cos x + i \sin x
$$

Raising $e^{ix}$ to the $n$th power where $n$ is an integer, we get from Euler's formula with $nx$ substituting for $x$
$$
(e^{ix})^n = e^{i(nx)} = \cos nx + i \sin nx
$$

Whenever $nx$ is an integer multiple of $2\pi$, we have
$$
\cos nx + i \sin nx = 1
$$

So
$$
e^{2\pi i \frac{k}{n}}
$$
is a root of 1 whenever $k/n = 0, 1, 2, \ldots$.

So the cube roots of unity are $1, e^{2\pi i/3}, e^{4\pi i/3}$. 

![img](figs/7_2.png)

While we can do this analytically, the idea is to use Newton's method to find these roots, and in the process, discover some rather perplexing behavior of Newton's method.

Newton's method for functions of complex variables - stability and basins of attraction. (30 points)

![img](figs/7_2B.png)

```python
# Newton function
def newton(z, f, fprime, max_iter=100, tol=1e-6):
    for i in range(max_iter):
        step = f(z)/fprime(z)
        if np.abs(step) < tol:
            return i, z
        z -= step
    return i, z
```

```python
# Function with 1 as cube root
def f(z):
    return z**3 - 1
```

```python
# Derivative of f
def fz(z):
    return 3*z**2
```

```python
# Record number of steps taken depending on starting points for Newton's method
def plot_newton(f, fz, n=200, extent=[-1,1,-1,1], cmap="hsv"):
    m = np.zeros((n,n))
    xmin, xmax, ymin, ymax = extent
    for r, x in enumerate(np.linspace(xmin, xmax, n)):
        for s, y in enumerate(np.linspace(ymin, ymax, n)):
            z = x + y*1j
            m[s, r] = newton(z, f, fz)[0]
    plt.imshow(m, cmap=cmap, extent=extent)
```

```python
# Rate of convergence depending on starting points
fig, ax = plt.subplots(figsize = (8,8))
plot_newton(f, fz);
```

```python
# Plot basins of attraction
def newton_basins(f, fz, n=200, extent=[-1,1,-1,1], cmap="jet"):
    root_count = 0
    roots = {}

    m = np.zeros((n,n))
    xmin, xmax, ymin, ymax = extent
    for r, x in enumerate(np.linspace(xmin, xmax, n)):
        for s, y in enumerate(np.linspace(ymin, ymax, n)):
            z = x + y*1j
            root = np.round(newton(z, f, fz)[1], 1)
            if not root in roots:
                roots[root] = root_count
                root_count += 1
            m[s, r] = roots[root]
    plt.imshow(m, cmap=cmap, extent=extent)   
```

```python
# Basins of attraction for newton's mathod
fig, ax = plt.subplots(figsize = (8,8))
newton_basins(f, fz);
```

Find solutions to $x^3 + 4x^2 -3 = x$. 

- Write a function to find brackets, assuming roots are always at least 1 unit apart and that the roots lie between -10 and 10
- For each bracket, find the enclosed root using
    - a bisection method
    - Newton-Raphson (no guarantee to stay within brackets)
- Use the end points of the bracket as starting points for the bisection methods and the midpoint for Newton-Raphson.
- Use the companion matrix and characteristic polynomial to find the solutions
- Plot the function and its roots (marked with a circle) in a window just large enough to contain all roots.

Use a tolerance of 1e-6.

![img](figs/7_4.png)

Notes:
- The loop invariant of the bisection method is the **bracket property**, i.e. $f(a)f(b)<0$
- With each step, find $c = (a+b)/2$ and $f(c)$, then update a/b to maintain the bracket property

```python
# Objective function
def f(x):
    return x**3 + 4*x**2 - x - 3
```

```python
# Derivative of objective function
def fx(x):
    return 3*x**2 + 8*x - 1
```

```python
# Bracket finder
def bracket(f, start, stop, step):
    brackets = []
    for a in np.arange(start, stop, step):
        b = a + step
        if f(a)*f(b) < 0:
            brackets.append([a,b])
    return brackets
```

```python
# Brackets for this problem
brackets = bracket(f, -10, 10, 1)
brackets
```

```python
# Bisection root finder to operate on one set of brackets
# Want to maintain bracket property f(a)f(b)<0, so set c accordingly
def bisection_roots(f, a, b, tol=1e-6):
    while np.abs(a-b) > tol:
        c = (a+b)/2
        if f(a)*f(c) < 0: # a/c different signs
            b = c
        else: # a/c same signs
            a = c
    return c
```

```python
# Find bisection roots
for i in brackets:
    a, b = i
    print(bisection_roots(f, a, b))
```

```python
# Newton-Raphson
def newton_raphson(f, fx, x, tol=1e-6):
    while np.abs(f(x)) > tol:
        x = x - f(x)/fx(x)
    return x
```

```python
# Find Newton-Raphson roots
for i in brackets:
    a, b = i
    x = (a+b)/2
    print(newton_raphson(f, fx, x))
```

Companion Matrix:
- Function: $x^3 + 4x^2 -x -3 = 0$
- Coefficient of the highest power is 1, so just flip the signs of the rest of the coefficients for the first row
- Filling out the rest of the matrix yields:
$$
\begin{bmatrix}
-4 & 1 & 3 \\
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{bmatrix}
$$

- Finding the roots of this polynomial is equivalent to finding the eigenvalues of the companion matrix

```python
# Companion matrix
comp = np.array([-4, 1, 3, 1, 0, 0, 0, 1, 0]).reshape(3,3)
```

```python
# Roots
roots = np.linalg.eigvals(comp)
roots
```

```python
# Plot
x = np.linspace(np.min(roots)-1, np.max(roots)+1, 100)
plt.plot(x, f(x))
plt.scatter(roots, np.zeros(len(roots)), c="red")
plt.axhline(0);
```
